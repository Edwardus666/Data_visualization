```python
import pandas
pd.set_option('mode.chained_assignment', None) 
```

```python
def count_school(page, data):
    counter = {}
    school = data
    
    for i in range(0,school['Institution'].size):  
            single_school = school['Institution'][i]
            counter[single_school] = 0
    #         print(single_school)

            if str(single_school) in str(page):
                if counter[single_school] == 1: #保证每一页出现的学校只会读取一次
                    break
                else:
                    school['rank'][i] += 1  
                    counter[single_school] = 1
            else:
                continue         
    return school
```

```python
# 第一次跑的时候用,以后都跑下一个cell
school['Institution'] = school['Institution'].str.replace(u'\xa0', ' ')
school['Institution'] = school['Institution'].str.strip()
# school['rank'].fillna(0, inplace = True) #仅当第一次run使用
```

```python
a = open('school.txt','r')
a = a.read()
school = pd.read_excel('琦申请研究生611.xlsx',sheet_name = 0)
school = count_school(a, school)
# 需要把正在打开的Excel关掉
school.to_excel('C:\\Users\\Edwar\\Desktop\\琦申请研究生611.xlsx')

# 这个可以看看
# https://www.kdnuggets.com/education/usa-canada.html#64c8ef28-fbe3-7be0-308b-eb38111d4eb6
```
